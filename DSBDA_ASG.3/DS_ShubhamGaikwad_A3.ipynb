```python
import pandas as pd
import numpy as np
import statistics
```

```python
df = pd.read_csv("loan_data_set.csv")
df.head()
```

```python
print("ApplicantIncome:-")
print(df.loc[:,"ApplicantIncome"].mean())
print("CoApplicantIncome:-")
print(df.loc[:,"CoapplicantIncome"].mean())
```

```python
print("ApplicantIncome Median:-")
print(df.loc[:,"ApplicantIncome"].median())
print("CoApplicantIncome Median:-")
print(df.loc[:,"CoapplicantIncome"].median())
```

```python
print("ApplicantIncome Minimum:-")
print(df.loc[:,"ApplicantIncome"].min())
print("CoApplicantIncome Minimum:-")
print(df.loc[:,"CoapplicantIncome"].min())
```

```python
print("ApplicantIncome maxium:-")
print(df.loc[:,"ApplicantIncome"].max())
print("CoApplicantIncome maxium:-")
print(df.loc[:,"CoapplicantIncome"].max())
```

```python
print("ApplicantIncome standard deviation:-")
print(df.loc[:,"ApplicantIncome"].std())
print("CoApplicantIncome standard deviation:-")
print(df.loc[:,"CoapplicantIncome"].std())
```

```python
print("Group By:-")
df.groupby(df['Gender']).ApplicantIncome.agg(['min','max','median','std'])
```
